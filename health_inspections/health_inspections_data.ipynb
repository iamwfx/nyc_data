```python
import pandas as pd
```

```python
health_inspections = pd.read_csv('DOHMH_New_York_City_Restaurant_Inspection_Results.csv')
```

```python
health_inspections.shape
```

```python
health_inspections.head()
```

```python
date = pd.to_datetime(health_inspections['INSPECTION DATE'], infer_datetime_format=True)
```

```python
health_inspections['date'] = date
```

```python
health_inspections_2017 = health_inspections[(health_inspections['date'] > '2017-01-01') & (health_inspections['date'] < '2018-01-01')]
```

```python
sample = health_inspections_2017[['BORO','BUILDING','STREET', 'ZIPCODE']]
```

```python
sample.shape
```

```python
sample.dropna(inplace=True)
sample.shape
```

```python
sample.ZIPCODE = sample.ZIPCODE.astype('int')
```

```python
sample.to_csv('sample_addresses.csv', index=False, header=['BoroughCode', 'HouseNumber', 'StreetName', 'ZipCode'])
```

```python
# read output
output = pd.read_csv('sample_output.csv')
```

```python
output.head()
```

```python
pd.eval(output.iloc[0]['lngLat'])
```

```python
output_latlng = output[~output['lngLat'].isna()]['lngLat'].apply(lambda x: pd.eval(x))
```

```python
outputnona = output[~output['lngLat'].isna()]
```

```python
outputnona['longitude'] = output_latlng.apply(lambda x: x[0])
outputnona['latitude'] = output_latlng.apply(lambda x: x[1])
```

```python
outputnona.columns
```

```python
outputnona.d()
```

```python
import cartoframes
cc = cartoframes.CartoContext()
```

```python
cc.write(outputnona[['HouseNumber', 'StreetName', 'BoroughCode','latitude','longitude']],
         'output_sample',
         overwrite=True,
        lnglat=('longitude','latitude'))
```

```python
cc.map(layers=cartoframes.Layer('output_sample'))
```
